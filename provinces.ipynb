```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = f"https://geokeo.com/database/state/ph/"
content_html = requests.get(url).content
soup = BeautifulSoup(content_html, "html.parser")
soup
```

```python
url = "https://geloky.com/geocoding/place/Cotabato+City+Philippines"
content_html = requests.get(url).content
soup = BeautifulSoup(content_html, "html.parser")
soup
```

```python
lat, long = str(soup).split("Soccsksargen place. Latitude: ")[1].split(". See where Cotabato City,")[0].split(" Longitude: ")
cotabato = {"PROVINCE": "COTABATO CITY", "LATITUDE": lat, "LONGITUDE": long}
cotabato
```

```python
url = r"https://geloky.com/geocoding/place/Metro+Manila+Philippines"
content_html = requests.get(url).content
soup = BeautifulSoup(content_html, "html.parser")
soup
```

```python
lat, long = str(soup).split("National Capital Region place. Latitude: ")[1].split(". See where Metro Manila,")[0].split(" Longitude: ")
mm = {"PROVINCE": "NATIONAL CAPITAL REGION", "LATITUDE": lat, "LONGITUDE": long}
mm
```

```python
all_provinces = []
for i in range(1, 6):
  url = f"https://geokeo.com/database/state/ph/{i}/"
  content_html = requests.get(url).content
  soup = BeautifulSoup(content_html, "html.parser")
  provinces = str(soup).split("<tbody><tr>\n")[1].split("</tr></tbody>")[0]
  provinces = provinces.replace('<a href="https://geokeo.com/database/state/ph/">', "").replace("</a>", "").replace("\n", '').replace('<td scope="col">', '')
  provinces = provinces.split("</tr><tr>")
  provinces = [p.split("</td>")[1:5] for p in provinces]
  provinces = [{"PROVINCE": p[0].upper(), "LATITUDE": p[2], "LONGITUDE": p[3]} for p in provinces]
  all_provinces += provinces
all_provinces.append(cotabato)
all_provinces.append(mm)
all_provinces
```

```python
all_provinces_df = pd.DataFrame.from_dict(all_provinces)
all_provinces_df
```

```python
all_provinces_df["LATITUDE"] = all_provinces_df["LATITUDE"].astype(np.float64)
all_provinces_df["LONGITUDE"] = all_provinces_df["LONGITUDE"].astype(np.float64)
all_provinces_df.dtypes
```
